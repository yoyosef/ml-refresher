```python
def encodeCategoricalColumns(X):
    # Step 1: Identify the categorical and numerical columns
    categorical_columns = ['State']
    numerical_columns = X.drop(columns=categorical_columns).columns.tolist()
    
    # Step 2: Apply OneHotEncoder to the categorical columns
    encoder = OneHotEncoder(drop='first', sparse_output=False)  # drop='first' to avoid the dummy trap
    
    # Fit and transform the categorical columns
    X_categorical_encoded = encoder.fit_transform(X[categorical_columns])
    
    # Convert the encoded result to a DataFrame
    encoded_column_names = encoder.get_feature_names_out(categorical_columns)
    X_categorical_encoded_df = pd.DataFrame(X_categorical_encoded, columns=encoded_column_names)
    
    # Step 3: Concatenate the encoded categorical columns with the numerical columns
    X_numerical = X[numerical_columns].reset_index(drop=True)
    X_final = pd.concat([X_numerical, X_categorical_encoded_df], axis=1)
    return X_final
```

```python
X_pd_final = encodeCategoricalColumns(X)
```

```python
X_final = X_pd_final.values[: , 1:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_final, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
y_pred = regressor.predict(X_test)
```
