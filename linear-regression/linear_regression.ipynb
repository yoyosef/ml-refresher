```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Student Scores is a csv filled with (hours, score) respectively.
# Predicting the score based on the number of hours that the student studied
dataset = pd.read_csv('studentscores.csv')
```

```python
# X values are the hours
X = dataset.iloc[ : ,   : 1 ].values
# Y values are the scores
Y = dataset.iloc[ : , 1 ].values
```

```python
from sklearn.model_selection import train_test_split
# 20% of the data is going to be used for testing
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 1/5, random_state = 13) 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
# Fitting linear regression model
regressor = regressor.fit(X_train, Y_train)
```

```python
# predictions
Y_pred = regressor.predict(X_test)
```

```python
# Plotting training values
plt.scatter(X_train , Y_train, color = 'red')
# Plotting linear model
plt.plot(X_train , regressor.predict(X_train), color ='blue')
```

```python
# Plotting predictions
plt.scatter(X_test , Y_test, color = 'red')
plt.plot(X_test , regressor.predict(X_test), color ='blue')
```
